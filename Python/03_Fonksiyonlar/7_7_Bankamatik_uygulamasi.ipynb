##### Para Çekme ve Para Yatırma fonksiyonları bulunmalıdır.

```python
Hafize_Hesap={
    'Ad':'Hafize Öğüt' ,
    'HesapNo':'5713071844',
    'bakiye':500,
    'EkHesap':1844,
    
}


Betul_Hesap={
    'Ad':'Bettt' ,
    'HesapNo':'3071844571',
    'bakiye':3070,
    'EkHesap':1840
} 
 
    
    

def ParaCek(hesap,miktar):
    print(f"Merhabalar {hesap['Ad']}")
    if (hesap['bakiye']>= miktar):
        
        hesap['bakiye']-=miktar
        
        print('paranızı alabilirsiniz')
        print(f"{hesap['HesapNo']} nolu ana hesabınızda {hesap['bakiye']} kalmıştır.")
        
    else:
        toplam=hesap['bakiye']+hesap['EkHesap']
        
        
        if (toplam >=miktar):
            ek_hesap_kullanimi=input("Ek hesap kullanmak ister misiniz?(e/h)")

            if (ek_hesap_kullanimi=='e'):

                hesap['bakiye']=0
                
                ek_hesap_kullanicalak_miktar=miktar-hesap['EkHesap']

                hesap['EkHesap']-=ek_hesap_kullanicalak_miktar
                print('paranızı alabilirsiniz') 
                print(f"{hesap['HesapNo']} nolu ek hesabınızda {hesap['bakiye']} bulunmaktadır.")

            else:
                print(f"{hesap['HesapNo']} nolu hesabınızda {hesap['bakiye']} bulunmaktadır.")
        else:
            print("Üzgünüz Bakiye Yetersiz")
            
def paraYatirma(hesap,miktar):
    print("Yatırılan paralar ana hesaba gitmektedir ")
    hesap['bakiye']+=miktar
    print(f"{hesap['HesapNo']} nolu ek hesabınızda {hesap['bakiye']} bulunmaktadır.")
    
    
    
    
def Bakiye_sorgulama(hesap):
    print(f"{hesap['HesapNo']} nolu hesabınızda {hesap['bakiye']} TL bulunmaktadır. ")
sifre='Rei-Ah Canım Sevgilim16/02'            
sayac=3
while True:
    kart_sifresi=input("Şifrenizi girmek ister misin? ")
    if (sifre==kart_sifresi):
        while True:
        
            print("""Başarılı bir şekilde giriş yapıldı.
            1-Para Çekme
            2-Para Yatırma
            3-Bakiye Sorgulama
            4-Kart iadesi
             
            """)

            secim=input("İşlem Numaranızı Seçiniz: ")

            if secim=='1':
                Cekilen_Tutar=int(input("Çekmek istediğiniz tutaru giriniz: "))
                ParaCek(Hafize_Hesap,Cekilen_Tutar)
            elif secim=='2':
                Yatirilan_Tutar=int(input("yatırmak istediğiniz tutaru giriniz: "))
                paraYatirma(Hafize_Hesap,Yatirilan_Tutar)
            elif secim=='3':
                Bakiye_sorgulama(Hafize_Hesap)
            elif secim=='4':
                print("Kartınız iade adiliyor.")
                break
                
        break 
    else:
        sayac-=1
    if sayac==0:
        print("Kartınızı alınız hesabınız bloke olmuştur....")
        break
    
```
