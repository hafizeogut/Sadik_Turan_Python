```python
import json
import os
class User:#kullanıcı adı parola mail
    def __init__(self,username,password,email):
        self.username=username
        self.password=password
        self.email=email

class UserRepository:
    def __init__(self):
        self.users=[]
        self.isLoggedIn=False#kullanıcı daha önceden sisteme kayıtlı mı?
        self.currentUser={}
           
        #load users from .json ifle
        #.json dosyasından kullanıcıları yükle
        self.load_Users()
        
    def load_Users(self):
        #load JSON verilerini Python verisine çevirirken, loads fonksiyonu veriyi parametre olarak alıyor
        if os.path.exists('users.json'):#Dosya mevcut dizinde var mı? Cevap:True False olacaktır
            
            with open('users.json','r',encoding='utf-8') as file:
                users=json.load(file)
                for user in users:
                    user=json.loads(user)#json bir ifadede python ifadesine değiştirildi.
                    #print(user['username'])
                    newUser=User(username=user['username'],password=user['password'],email=user['email'])
                    self.users.append(newUser)
            print(self.users)
        
        
    def register(self,user:User):#kullanıcı oluşturma
        self.users.append(user)#UserRepository içerisinde tanımlanan users a ulaşıldı ve user bilgileri eklendi
        self.save_to_File()#kayıt işlemi:bütün listeyi dosyaya ekleme işlemi yapıldı.
        
        print('Kullanıcı oluşturuldu')
    
    def login(self,username,password):#giriş ekranı oluşturma
        
        for user in self.users:
            if user.username==username and user.password==password:
                self.isLoggedIn=True
                self.currentUser=user
                print('login yapıldı.')
                break
    def  logout(self):
        self.isLoggedIn=False
        self.currentUser={}
        print('Çıkış yapıldı')
        
    def identity_kimlik(self):
        if self.isLoggedIn:
            print(f'username: {self.currentUser.username}')
            
        else:
            print('giriş yapılmadı.')
            
    
    def save_to_File(self):#json olarak ver tabanına kayıt edilecek
        liste=[]
        for user in self.users:#user değişkeni ilse users listesinin içerisinde gezinildi.
            liste.append(json.dumps(user.__dict__))#user sınıfı bir sözlük yapısına çevrildi.
            
        
        with open('users.json','w')as file:
            #json.dumps(self.users,file)#kayıt edilecek objeyi kayıt edilebilecek formata getirme işlemi yapıldı.
            
            json.dump(liste,file)#self.users objesi  file dosyasına kaydedildi.
            #json içerisine göndereceğim argümanları liste sözlük gibi bir formatta olması beklenir.
            #Bu yüzden elimizdeki bilgileri sözlük formatına çevirmek gerekmektedir.
            
    
    
repository=UserRepository()

while True:
    print('Menü'.center(50,'*'))#50 karakte içerisinde * ile ortalanmış bir menü oluşturuldu
    secim=input('1- Register\n2-Login\n3-Lohout\n4-identity\n5-Exit\nSeçiminiz: ')
    if secim=='5':
        break
        
    else:
        if secim=='1':
            #register:kayıt
            username=input('username: ')
            password=input('password: ')
            email=input('mail: ')
            
            user=User(username=username,password=password,email=email)#User classı user değişkenin parametreleri ile birlikte aktarıldıç 
            repository.register(user)#register metoduna UserRepository() classından ulaşıldı.
            
            print(repository.users)
            
        elif secim=='2':
            #login
            
            if repository.isLoggedIn:
                print("Zaten giriş yaptınız")
            
            else:
                username=input('username: ')
                password=input('password: ')
                repository.login(username,password)
        
        elif secim=='3':
            #logout
            repository.logout()
        
        elif secim=='4':
            repository.identity_kimlik()
        else:
            print("yanlış seçim")
        
        
```
