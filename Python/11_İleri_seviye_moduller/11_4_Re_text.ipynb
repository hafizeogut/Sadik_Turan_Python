```python
import re
string=""""
[]- Köeşeli parantezler arasında yazılan bütün karakterler aranır.

[abc]=> a     :1 match
        b     :2 match
        Python:No matches
        
        
    [a-e] =>[abcde]
    [1-5] =>[12345]
    [0-39]=>[01239]
    
    [^abc] =>abc dışındaki karakterler.
    [^0-9] =>rakam olmayan karakterler.
"""
re.findall("[abc]",string)
```

```python
re.findall("[saat]",string)
```

```python
re.findall("[a-e]",string)
```

```python
re.findall("[1-5]",string)
```

```python
re.findall("[^abc]",string)#^ haricindeki demektir.
```

```python
string1="""
    -Her hangi bir tek karakteri belirtir.
    
        ..=>a    :No match
            ab   :1 match
            abc  :1 match
            abcd :2 matches
            
"""

re.findall("...",string1)

```

```python
re.findall("py..on",string1)
```

```python
str2="""
    ^ - Belirtilen karakter ile başlıyor mu ?
        ^a=> a  : 1 match
             abc: 1 match
             bac: No match

"""
re.findall("^",str2)
```

```python
str2="""
    $ - Belirtilen karakter ile bitiyor mu ?
    ^$=>     a     : 1 match
             lambda: 1 match
             Python: No match

"""
re.findall("karakter$",str2)
```
